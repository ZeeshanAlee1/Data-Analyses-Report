```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
path="./Super_market_sales"
os.listdir(path)
```

```python
sales = pd.read_csv(r"C:\Users\Amir Computer Shop\Desktop\Sales_Analysis\Super_market_sales\supermarket_sales.csv" , encoding='latin1')
sales.head(1000)
```

```python
sales.info()
```

```python
sales = sales.dropna(how='all')
sales.head()
```

```python
sales['Quantity'] = pd.to_numeric(sales['Quantity'])
sales['Unit price'] = pd.to_numeric(sales['Unit price'])
```

```python
sales = sales[sales['Date'].str[0:1]!='or']
```

```python
sales['Month'] = sales['Date'].str[0:1]
#sales['Month'] = sales['Month'].astype('int32')
sales.head(1500)
```

```python
sales['Sales'] = sales ['Quantity'].astype('int') * sales['Unit price'].astype('float')
sales.head()
```

```python
sns.set(style="darkgrid") 
genderCount  = sns.countplot(x="Gender", data =sales).set_title("Gender_Count")
```

```python
sns.boxplot(x="Branch", y = "Rating" ,data =sales).set_title("Ratings by Branch") 
```

```python
type(sales['Date'])
sales['Date'] = pd.to_datetime(sales['Date'])
sales['day'] = (sales['Date']).dt.day
sales['month'] = (sales['Date']).dt.month
sales['year'] = (sales['Date']).dt.year
sales['Time'] = pd.to_datetime(sales['Time'])
sales['Hour'] = (sales['Time']).dt.hour 
sales['Hour'].nunique()
sales['Hour'].unique()
```

```python
genderCount  = sns.lineplot(x="Hour",  y = 'Quantity',data =sales).set_title("Product Sales per Hour")
```

```python
genderCount  = sns.relplot(x="Hour",  y = 'Quantity', col= 'month' , row= 'Branch', kind="line", hue="Gender", style="Gender", data =sales)
```
